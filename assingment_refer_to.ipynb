```python
languages_count = {}
with open(r'c:\Users\Admin\Documents\data.txt') as csv_file:
    csv_reader= csv.DictReader(csv_file)
    lgs_count = {}
    for data in csv_reader:
        for lgs in data['LanguagesWorkedWith'].split(';'):
            lgs_count.setdefault(lgs,0)
            lgs_count[lgs] +=1
print(lgs_count)
```

```python
languages = {'Languages': list(lgs_count.keys()), 'popularity': list(lgs_count.values())}
# languages
df = pd.DataFrame(languages)
df
```

```python
# sorting top 15 use language 
data = sorted(lgs_count.items(), key = lambda kv:kv[1], reverse= True)
new_data = data[:16]
Popularity = [new_data[i][1] for i in range(len(new_data))]
Languages = [new_data[i][0] for i in range(len(new_data))]
# creating top 15 data frame
df = pd.DataFrame(zip(Languages, Popularity), columns=['Languages', 'Popularity'])
df
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(20,20))
x = df.Languages
y = df.Popularity
ax = plt.axes()
ax.barh(x, y)
ax.set_title('simple survey on popularity of programming language')
ax.set_xlabel('Popularity', fontsize=15, fontstyle='italic')
ax.set_ylabel('Programming language', fontsize=15, fontstyle= 'italic')
plt.xticks(fontstyle='italic')
plt.yticks(fontstyle='italic')
plt.gca().invert_yaxis()
plt.savefig('Popularity.jpg')
plt.tight_layout()
plt.show()
```

```python
# simple pie chart
plt.style.use('fivethirtyeight')

slices = Popularity[:5]
labels = Languages[:5]
explode = [0.1, 0, 0, 0.1, 0]
colors = ['blue', 'red', 'green', 'orange', 'white']
plt.pie(slices, labels=labels,colors=colors,explode=explode,shadow=True,startangle=180, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})
plt.savefig('Popularity.jpg')
plt.tight_layout()
plt.show()
```
